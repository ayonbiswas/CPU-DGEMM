```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import re
```

```python
%matplotlib inline
#sns.set_theme()
```

```python
plt.rcParams['figure.figsize'] = [8, 6]
plt.rcParams['figure.dpi'] = 100 # 200 e.g. is really fine, but slower
```

```python
sns.set_style("whitegrid")
```

```python
folder = './L1_blocks/'
block_files = os.listdir(folder)
MR = 4
name = "Performance by L1-block size"
```

```python
plt.Figure()
for i in range(len(block_files)):
    data = np.loadtxt(os.path.join(folder,block_files[i]))
    KC = int(re.split("_",block_files[i],2)[-1].split(".")[0])
    block_size = (KC*MR*8)//1024
    print(block_files[i],block_size)
    plt.plot(data[:,0], data[:,1],label='{} KB'.format(block_size))
plt.title("{} ".format(name), size=17)
plt.ylabel("Performance(GFLOPS)", size=14)
plt.xlabel("Matrix dimension", size=14)
# plt.xscale("log")
plt.grid(True)
plt.legend()
# plt.savefig("./cse260plots/L1_blocks.png", dpi = 150, bbox_inches='tight')
# plt.close()
```
